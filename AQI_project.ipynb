```python
# Cell 1: Install (if needed) + Import
import requests
import os
from datetime import datetime
from dotenv import load_dotenv

# Load .env file
load_dotenv()

# ================== YOUR TOKEN ==================
TOKEN = os.getenv("AQICN_TOKEN")

if not TOKEN:
    print("X Token not found! Please check your .env file")
else:
    print(":) Token loaded successfully")
# ================================================

def get_aqi(city="islamabad"):
    url = f"https://api.waqi.info/feed/{city}/?token={TOKEN}"
    
    print(f"Checking AQI for: {city.title()}")
    print(f"Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        if data.get('status') == 'ok':
            aqi = data['data']['aqi']
            city_name = data['data']['city']['name']
            
            print(f":) SUCCESS!")
            print(f"Location : {city_name}")
            print(f"Current AQI : {aqi}")
            
            # Color coding
            if aqi <= 50:   color = "🟢 Good"
            elif aqi <= 100: color = "🟡 Moderate"
            elif aqi <= 150: color = "🟠 Unhealthy for Sensitive"
            else:            color = "🔴 Unhealthy/Hazardous"
                
            print(f"Air Quality: {color}")
            return aqi
        else:
            print(":( API Response Error:", data.get('data'))
    else:
        print(":( HTTP Error:", response.status_code)
        print(response.text)

# Run the function
get_aqi("islamabad")
```
